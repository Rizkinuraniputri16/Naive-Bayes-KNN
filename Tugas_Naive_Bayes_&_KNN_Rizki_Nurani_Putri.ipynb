Setiap baris dalam dataset mewakili satu pasien, dan kolom-kolomnya adalah atribut-atribut yang diamati pada pasien-pasien tersebut. Berikut adalah penjelasan dari setiap variabelnya:

1. Age: Umur pasien dalam tahun.
2. Number.of.sexual.partners: Jumlah pasangan yang pernah berhubungan dengan pasien.
3. First.sexual.intercourse: Umur pertama kali pasien berhubungan badan.
4. Num.of.pregnancies: Jumlah kehamilan yang pernah dialami oleh pasien.
5. Smokes: Status apakah pasien merokok atau tidak (biasanya berupa data kategorikal seperti "Ya" atau "Tidak").
6. Smokes..years.: Lama waktu pasien merokok, diukur dalam satuan tahun.
7. Smokes..packs.year.: Jumlah kumulatif paparan rokok yang dihitung dalam satuan pack-year (jumlah bungkus rokok yang dihisap per hari dikalikan dengan jumlah tahun merokok).
8. Hormonal.Contraceptives: Status penggunaan kontrasepsi hormonal oleh pasien (biasanya berupa data kategorikal seperti "Ya" atau "Tidak").
9. Hormonal.Contraceptives..years.: Lama waktu pasien menggunakan kontrasepsi hormonal, diukur dalam satuan tahun.
10. IUD: Status penggunaan alat kontrasepsi dalam rahim (Intrauterine Device) oleh pasien (biasanya berupa data kategorikal seperti "Ya" atau "Tidak").
11. IUD..years.: Lama waktu pasien menggunakan IUD, diukur dalam satuan tahun.
12. STDs: Status apakah pasien pernah atau sedang mengalami infeksi menular seksual (Sexually Transmitted Diseases), biasanya berupa data kategorikal "Ya" atau "Tidak".
13. STDs..number.: Jumlah total infeksi menular seksual yang pernah diderita pasien (termasuk episode berulang).
14. STDs.condylomatosis: Keberadaan kondiloma akuminata (kutil kelamin) yang disebabkan oleh HPV tipe 6/11, sering muncul sebagai lesi vegetatif di area genital.
15. STDs.cervical.condylomatosis: Kondiloma akuminata yang terletak di serviks (leher rahim).
16. STDs.vaginal.condylomatosis: Kutil kelamin yang muncul di dinding vagina.
17. STDs.vulvo.perineal.condylomatosis: Kutil di area vulva dan perineum (daerah antara vagina/anus).
18. STDs.syphilis: Infeksi bakteri Treponema pallidum dengan prevalensi global 45,4 juta kasus, ditandai ulkus genital dan komplikasi sistemik jika tidak diobati.
19. STDs.pelvic.inflammatory.disease: Radang panggul (PID) akibat infeksi bakteri seperti Chlamydia atau Gonorrhea, berisiko menyebabkan infertilitas.
20. STDs.genital.herpes: Infeksi HSV-2 dengan gejala vesikel nyeri dan rekurensi periodik.
21. STDs.molluscum.contagiosum: Infeksi virus Molluscipoxvirus menyebabkan papul umbilikasi di area genital.
23. STDs.AIDS: Stadium akhir infeksi HIV dengan CD4 <200 sel/mm³ atau adanya infeksi oportunistik.
24. STDs.HIV: Virus penyebab defisiensi imun dengan risiko transmisi 0.05-1.7% per aktivitas seksual.
25. STDs.Hepatitis.B: Infeksi virus hepatotropik yang dapat dicegah dengan vaksin, prevalensi 356 juta kasus global.
26. STDs.HPV: Virus penyebab 800 juta infeksi global, terkait dengan 99% kasus kanker serviks (tipe 16/18).
27. STDs..Number.of.diagnosis: Jumlah episode diagnosis IMS yang pernah diterima pasien.
28. Dx.Cancer: Diagnosis kanker (misal: karsinoma serviks) berdasarkan hasil histopatologi.
29. Dx.CIN: Cervical Intraepithelial Neoplasia (lesi prakanker serviks) dengan gradasi CIN1-CIN3.
30. Dx.HPV: Konfirmasi infeksi HPV melalui PCR atau hibridisasi DNA.
31. Dx: Diagnosis umum penyakit berdasarkan anamnesis, pemeriksaan fisik, dan penunjang.
32. Hinselmann: Ujian kolposkopi menggunakan asam asetat 3-5% untuk mendeteksi lesi serviks.
33. Schiller: Tes dengan larutan iodin untuk mengidentifikasi epitel serviks abnormal.
34. Citology: Pemeriksaan sitologi (Pap smear) untuk mendeteksi sel abnormal atau keganasan.
35. Class_label: Variabel target atau label yang menunjukkan apakah pasien memiliki kanker serviks atau tidak. Label ini umumnya bernilai 1 untuk kasus positif (pasien memiliki kanker serviks) dan 0 untuk kasus negatif (pasien tidak memiliki kanker serviks).

Dataset ini memberikan informasi penting tentang karakteristik klinis pasien dan dapat digunakan untuk memprediksi atau menganalisis faktor-faktor risiko yang terkait dengan kanker serviks.

```python
#memisahkan data menjadi variabel x dan y
X=cervical.drop(['class_label'], axis=1)
y=cervical['class_label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler

X_train = MinMaxScaler().fit_transform(X_train)
X_test = MinMaxScaler().fit_transform(X_test)

print(X_train)
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
# fit train set pakai Gaussian Naive Bayes
nb.fit(X_train, y_train)
```

```python
# compute the accuracy of train set
nb.score(X_train, y_train)
```

```python
classif_name = ['NB']
y_predict = {}
y_predict['NB'] = nb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score  # Import accuracy_score

# Define classif_name before using it in the loop
classif_name = ['NB']
y_predict = {}
y_predict['NB'] = nb.predict(X_test)

# Now you can iterate through classif_name
for name in classif_name:
    print('{0} accuracy = {1:.4f}'.format(name, accuracy_score(y_test, y_predict[name])))
```

Output 0.2239 yang dihasilkan NB accuracy menunjukkan tingkat akurasi dari model klasifikasi yang dilatih. Dalam konteks ini, angka 0.2239 berarti bahwa model berhasil memprediksi dengan benar sekitar 22,39% dari semua sampel pada data uji (y_test) yang telah digunakan.

## KNN

```python
#Menguji parameter yang paling optimum untuk model

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)



# Tentukan parameter yang ingin diuji
parameters = {
    'hidden_layer_sizes': [(50,), (100,), (50, 30), (100, 50)],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'max_iter': [100, 200, 300]
}

# Buat model MLPClassifier
clf = MLPClassifier(random_state=100)

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(clf, parameters, cv=cv_strategy)
# cv merupakan jumlah lipatan validasi silang (cross-validation)

# Lakukan pencarian grid pada data training
grid_search.fit(X, y)
```

```python
# Tampilkan parameter terbaik
print("Parameter terbaik:", grid_search.best_params_)
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(hidden_layer_sizes=(100, 50),learning_rate_init=0.1,max_iter=100)

clf.fit(X_train, y_train)
```

Dalam kasus ini, model MLPClassifier dibuat dengan parameter-parameter tertentu yang telah ditentukan secara manual. Parameter-parameter ini adalah:

hidden_layer_sizes=(100, 50): Ini adalah tuple yang menunjukkan arsitektur dari jaringan saraf. (100, 50) berarti terdapat dua hidden layers berisi 100 dan 50 neuron.

learning_rate_init=0.1: Parameter ini adalah tingkat pembelajaran awal. Ini menunjukkan seberapa besar model akan menyesuaikan bobotnya berdasarkan kesalahan pada setiap iterasi.

max_iter=100: Parameter ini menunjukkan jumlah maksimum iterasi yang akan dilakukan saat melatih model. Ini mengontrol seberapa banyak iterasi atau epoch yang akan dilakukan selama proses pelatihan.

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

Hasil dari laporan klasifikasi tersebut memberikan ringkasan performa model Neural Network pada data uji (cervical) dengan menggunakan berbagai metrik evaluasi seperti precision, recall, dan f1-score. Berikut adalah interpretasi dari setiap metrik yang tercantum dalam laporan:

Precision: 0 (Kelas Negatif): Dari semua prediksi yang dilakukan sebagai kelas negatif, sekitar 97% di antaranya adalah benar-benar kelas negatif. 1 (Kelas Positif): Sekitar 33% dari semua prediksi yang dilakukan sebagai kelas positif adalah benar.

Recall: 0 (Kelas Negatif): Dari semua sampel yang sebenarnya adalah kelas negatif, model mampu mengenali sekitar 90% dari mereka. 1 (Kelas Positif): Dari semua sampel yang sebenarnya adalah kelas positif, model hanya dapat mengenali sekitar 60% dari mereka.

F1-Score: F1-score adalah ukuran rata-rata harmonis dari precision dan recall. F1-score untuk kelas 0 adalah 0.93, sedangkan untuk kelas 1 adalah 0.43. F1-score semakin tinggi semakin baik, karena mencoba untuk menggabungkan keseimbangan antara precision dan recall.

Accuracy: Akurasi model adalah sekitar 88%, yang mengindikasikan seberapa baik model dapat memprediksi dengan benar secara keseluruhan dari seluruh sampel.

Interpretasi: Meskipun akurasi secara keseluruhan adalah 88%, melihat pada precission untuk kelas 1 (0.33) menunjukkan bahwa model memiliki performa yang lebih rendah dalam mengidentifikasi kasus positif (1). Terlihat bahwa model memiliki kemampuan yang baik dalam memprediksi kasus negatif (0) tetapi kurang baik dalam memprediksi kasus positif (1). Interpretasi ini penting tergantung pada konteks aplikasi, karena keseimbangan antara kelas 0 dan 1 bisa menjadi kunci dalam pengambilan keputusan yang dihasilkan oleh model.
